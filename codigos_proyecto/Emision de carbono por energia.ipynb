```python
#me estaba generando problemas los espacios por lo tanto, se realizo lo siguiente para eliminarlos:
lista_nombres=[]
for name in df_T.columns:
    lista_nombres.append(name.strip())
```

```python
clean_names = [name.strip() for name in df_T.columns] #list comprehension
df_T.columns = clean_names
```

```python
#para hacer un top de paises es necesario eliminar los totales, y conjunto de paises que interfieren en el analisis.segunda forma de hacerlo.
top10_emite_mas_CO2 = df_T.sort_values("Años",ascending=False).iloc[:11,:]
eliminar = [x for x in top10_emite_mas_CO2.columns if ("total") in x.lower()]
eliminar
```

```python
c =[]
for columna in top10_emite_mas_CO2.columns:
    if "total" in columna.lower():
        c.append(columna)
c
```

```python
#agrupe todos los valores que queria eliminar por que no eran necesarios para el analisis.
elim=c+["of which: OECD","Total Asia Pacific","Non-OECD","Total World","Other Asia Pacific","European Union #", "Años"]
elim
```

```python
#se eleiminaron 
top10_emite_mas_CO2.drop(columns=elim, inplace=True)
```

```python
#tabla con los valores filtrados.
top10_emite_mas_CO2.columns
```

```python
#se realiza un promedio de las emision de los paises de tomando las emisiones con eje central
promedio=top10_emite_mas_CO2.mean()
promedio
```

```python
#se toma realiza un listado de paises con sus correspondientes emisiones 
top10_emite_mas_CO2= promedio.reset_index()
top10_emite_mas_CO2.columns=["pais", "Emision De CO2 (Mt)"]
```

```python
#se toman los 10 paises que mas emiten CO2 y se graficaran
top10_emite_mas_CO2= top10_emite_mas_CO2.sort_values("Emision De CO2 (Mt)", ascending=False).head(10)
top10_emite_mas_CO2
```

```python
#grafica correspondiente:
sns.barplot(data=top10_emite_mas_CO2, x="Emision De CO2 (Mt)", y="pais")
plt.title("Top paises con mas emisiones (promedio) en los ultimos 10 años")
```

```python
#se realiza el mismo procedimiento para el top de los que menos CO2 Emiten, teniendo cuidado con el uso de funciones:
top10_emite_menos_CO2 = df_T.sort_values("Años",ascending=False).iloc[:11,:]
top10_emite_menos_CO2 
```

```python
eliminar = [x for x in top10_emite_menos_CO2.columns if "total" in x.lower()]
eliminar
```

```python
d =[]
for columna in top10_emite_menos_CO2.columns:
    if "total" in columna.lower():
        d.append(columna)
d
```

```python
d=d+["of which: OECD","Total Asia Pacific","Non-OECD","Total World","Other Asia Pacific","European Union #", "Años","USSR"]
d
```
