```python
#me estaba generando problemas los espacios por lo tanto, se realizo lo siguiente para eliminarlos:
lista_nombres=[]
for name in df_T.columns:
    lista_nombres.append(name.strip())
```

```python
clean_names = [name.strip() for name in df_T.columns] #list comprehension
df_T.columns = clean_names
```

```python
#para hacer un top de paises es necesario eliminar los totales, y conjunto de paises que interfieren en el analisis.segunda forma de hacerlo.
top10_produce_mas_oil = df_T.sort_values("Años",ascending=False).iloc[:11,:]
eliminar = [x for x in top10_produce_mas_oil.columns if ("total") in x.lower()]
eliminar
```

```python
c =[]
for columna in top10_produce_mas_oil.columns:
    if "total" in columna.lower():
        c.append(columna)
c
```

```python
elim=c+["of which: OECD","Total Asia Pacific","Non-OECD","Total World","Other Asia Pacific","European Union #", "Años","OPEC",'Non-OPEC']
elim
```

```python
#se eleiminaron 
top10_produce_mas_oil.drop(columns=elim, inplace=True)
```

```python
#tabla con los valores filtrados.
top10_produce_mas_oil.columns
```

```python
#se realiza un promedio de la produccion de crudo de los paises de todos los años correspondientes.
promedio=top10_produce_mas_oil.mean()
promedio
```

```python
#se toma realiza un listado de paises con sus correspondientes producciones de crudo
top10_produce_mas_oil= promedio.reset_index()
top10_produce_mas_oil.columns=["pais", "Produccion De Crudo(Mb/d)"]
```

```python
#se toman los 10 paises que mas producen crudoy se graficaran
top10_produce_mas_oil= top10_produce_mas_oil.sort_values("Produccion De Crudo(Mb/d)", ascending=False).head(10)
top10_produce_mas_oil
```

```python
#grafica correspondiente:
sns.barplot(data=top10_produce_mas_oil, x="Produccion De Crudo(Mb/d)", y="pais")
plt.title("Top paises con mas produccion de crudo (promedio) en los ultimos 10 años")
```

```python
#se realiza el mismo procedimiento para el top de los que menos producen, teniendo cuidado con el uso de funciones:
top10_produce_menos_oil = df_T.sort_values("Años",ascending=False).iloc[:11,:]
top10_produce_menos_oil
```

```python
eliminar = [x for x in top10_produce_menos_oil.columns if "total" in x.lower()]
eliminar
```

```python
d =[]
for columna in top10_produce_menos_oil.columns:
    if "total" in columna.lower():
        d.append(columna)
d
```

```python
d=d+["of which: OECD","Total Asia Pacific","Non-OECD","Total World","Other Asia Pacific","European Union #", "Años","USSR"]
d
```

```python
top10_produce_menos_oil.drop(columns=d, inplace=True)
```

```python
promedio=top10_produce_menos_oil.mean()
promedio
```

```python
top10_produce_menos_oil= promedio.reset_index()
top10_produce_menos_oil.columns=["Pais", "Produccion De Crudo(Mb/d)"]
```

```python
top10_produce_menos_oil= top10_produce_menos_oil.sort_values("Produccion De Crudo(Mb/d)", ascending=True).head(10)
top10_produce_menos_oil
```

```python
sns.barplot(data=top10_produce_menos_oil, x="Produccion De Crudo(Mb/d)", y="Pais",color="green")
plt.title("Top paises con menos produccion de crudo (promedio) en los ultimos 10 años")
```
